```python
from PIL import Image, ImageDraw, ImageFont

# create an image
img = Image.open('title_bg2.png')

# get a font
fnt = ImageFont.truetype("arial.ttf", 50, encoding="unic")
# get a drawing context
d = ImageDraw.Draw(img)

# draw multiline text
d.multiline_text((10, 600), "Chemical\nGroup\n Analysis", font=fnt, fill=(255, 255, 255 ))

img.show()
img.save('title_new.png')
```

```python
from docx import Document
from docx.shared import Inches

document = Document()

p = document.add_paragraph()
r = p.add_run()
#r.add_text('')
r.add_picture('title_new.png',width=Inches(6), height=Inches(9))
r.add_text(' do you like it?')

document.save('demo.docx')
```

```python
import pandas as pd
```

```python
csv_data=pd.read_excel('excels\\Cleanery - handwash.xlsx',sheet_name='Graph&Weightings',skiprows=range(1, 55))
```

```python
csv_data.fillna(' ')
```

```python
document = Document()
df=csv_data.
```

```python
t = document.add_table(csv_data.shape[0]+1, csv_data.shape[1])
```

```python
# add the header rows.
for j in range(df.shape[-1]):
    t.cell(0,j).text = df.columns[j]
```

```python
# add the rest of the data frame
for i in range(df.shape[0]):
    for j in range(df.shape[-1]):
        t.cell(i+1,j).text = str(df.values[i,j])

# save the doc
document.save('./test.docx')
```
